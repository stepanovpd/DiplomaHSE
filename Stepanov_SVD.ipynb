```python
matrix=pd.pivot_table(data=Rt1_df[['Phone_new', 'TK', 'Quant']], values='Quant', index='Phone_new',columns='TK')
matrix = matrix.fillna(0)
matrix
```

```python
def hybrid_recommendations(userId, productId):
    top_five=matrix.corrwith(matrix[productId]).sort_values(ascending=False).head(5)
    
    est_rating=[]
    for x in list(top_five.index):
        if str(top_five[x])!='nan':
            est_rating.append(svd.predict(userId, iid=x, r_ui=None).est)
           
    return pd.DataFrame({'productId':list(top_five.index)[:len(est_rating)], 'estimated_rating':est_rating}).sort_values(by='estimated_rating', ascending=False).reset_index(drop=True)

def all_hybrid_recommendations(userId, TKname):
    
    top_five=matrix.corrwith(matrix[TKname])
    
    est_rating=[]
    for x in list(top_five.index):
        est_rating.append(svd.predict(userId, iid=x, r_ui=None).est)
           
    return pd.DataFrame({'productId':list(top_five.index)[:len(est_rating)], 'estimated_rating':est_rating}).sort_values(by='estimated_rating', ascending=False).reset_index(drop=True)

def product_mapping(recommender_df, Rt_df): 
    rows_list = []
    rows_list2 = []

    for x in recommender_df['productId']:
        cut_row = Rt_df.loc[Rt_df['TK'] == x]
        cut_row = cut_row['TK'][0:1].values
        rows_list.append(cut_row[0])

    for i in recommender_df['estimated_rating']:
        rows_list2.append(i)

    cut_dict = {"product": rows_list, "estimated_rating": rows_list2}
    cut_df = pd.DataFrame(cut_dict)
    
    return cut_df
```

```python
find_product1 = hybrid_recommendations(Rt1_df['Phone_new'][len(Rt1_df)//10], 'Стиральные машины')
find_all_product1 = all_hybrid_recommendations(Rt1_df['Phone_new'][len(Rt1_df)//10], 'Стиральные машины')
```

```python
product_mapping(find_product1, Rt1_df)
```

```python
product_mapping(find_all_product1, Rt1_df).head(5)
```
